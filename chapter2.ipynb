```python
WEEKDAY = {0: '月', 1: '火', 2: '水', 3: '木', 4: '金', 5: '土', 6: '日'}
```

```python
WEEKDAY
```

```python
for num, name in WEEKDAY.items():
    print('{}: {}'.format(num, name))
```

```python
WEEKDAY.items()
```

```python
def get_day_of_week(date_str):
    dt =datetime.strptime(date_str, '%Y/%m/%d')
    return WEEKDAY[dt.weekday()]
```

```python
get_day_of_week('2019/1/11')
```
